```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://docs.google.com/spreadsheets/d/1VMaw2oCn0ABitd-alLAEsEhGS1Je2UFNLu76TKrIH7w/gviz/tq?tqx=out:csv&sheet=Raw_Medical_Data_for_day1'
df = pd.read_csv(url, index_col = 0)
df.head()
```

```python
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(df['Income'], ax = ax, stat='probability');
```

```python
# Find the mean
mean = df['Income'].mean()
mean
```

```python
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(df['Income'], ax = ax, stat='probability')
ax.axvline(mean);
```

```python
# Change line color to red
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(df['Income'], ax = ax, stat='probability')
ax.axvline(mean, color = 'red');
```

```python
# Add legend and title
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(df['Income'] ,ax=ax, stat='probability')
ax.axvline(mean, color = 'red', label='Mean')
ax.set_title('Income')
ax.legend();
```

```python
## Add value for mean with f string
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(df['Income'] ,ax=ax,stat='probability')
ax.axvline(mean,color = 'red', label=f'Mean = {mean:,.2f}')
ax.set_title('Income')
ax.legend();
```

```python
col = 'Income'
```

```python
feature = df[col]
mean = feature.mean()
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(feature ,ax=ax,stat='probability')
ax.axvline(mean,color = 'red', label=f'Mean = {mean:,.2f}',)
ax.set_title(f'{col}')
ax.legend();
```

```python
col = 'Income'
feature = df[col]
mean = feature.mean()
median = feature.median()
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(feature ,ax=ax,stat='probability')
ax.axvline(mean, color = 'red',label=f'Mean = {mean:,.2f}')
ax.axvline(median, ls='--', color = 'green', label = f'Median = {median:,.2f}')
ax.set_title(f'{col}')
ax.legend();
```

```python
col = 'Income'
feature = df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(feature ,ax=ax,stat='probability')
ax.axvline(plus_one_std, color = 'black',label=f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}')
ax.set_title(f'{col}')
ax.legend();
```

```python
col = 'Income'
feature = df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(feature ,ax=ax,stat='probability')
ax.axvline(plus_one_std, color = 'black',label=f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)
ax.set_title(f'{col}')
ax.legend();
```
